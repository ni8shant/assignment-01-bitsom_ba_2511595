```
from sentence_transformers import SentenceTransformer, util
import numpy as np
import matplotlib.pyplot as plt

# Initialize model
model = SentenceTransformer('all-MiniLM-L6-v2')

# 10 sentences across 3 topics
sentences = [
  "Virat Kohli is an excellent cricket batsman",
  "The bowler took three wickets in one over",
  "India defeated Australia in the Test match",
  "Cooking pasta requires boiling water",
  "The recipe calls for fresh herbs and spices",
  "A well-prepared sauce enhances the dish",
  "Cybersecurity is critical for protecting data",
  "Hackers use various techniques for infiltration",
  "Encryption protects sensitive information online",
  "Web applications must validate user inputs"
]

# Generate embeddings
embeddings = model.encode(sentences)

# Compute similarity matrix
similarity_matrix = util.pytorch_cos_sim(embeddings, embeddings)

print("Similarity Matrix (10x10):")
print(similarity_matrix)

# Find most similar sentences to query
query = "The bowler took three wickets in one over"
query_embedding = model.encode(query)
similarities = util.pytorch_cos_sim(query_embedding, embeddings)[0]
top_indices = np.argsort(similarities)[-3:]\[::-1]

print(f"\nQuery: {query}")
for idx in top_indices[:2]:
  print(f"  {sentences[idx]} (similarity: {similarities[idx]:.4f})")
```
